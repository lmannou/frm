Bernouilli

```python
from scipy.stats import bernoulli
# Create an array of values for the x-axis
x = np.arange(2)

# Plot the Bernoulli distribution with p = 0.8
plt.bar(x, bernoulli.pmf(x, 0.8), color=['red', 'blue'], label=['failure', 'success'])
plt.xticks(x, ['0', '1'])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.legend()
plt.title('Bernoulli Distribution with p = 0.8')
plt.show()
```

```python
from scipy.stats import bernoulli
# Create an array of values for the x-axis
x = np.arange(2)

# Plot the Bernoulli distribution with p = 0.8
plt.bar(x, bernoulli.cdf(x, 0.8), color=['red', 'blue'], label=['failure', 'success'])
plt.xticks(x, ['0', '1'])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.legend()
plt.title('Bernoulli Distribution with p = 0.8')
plt.show()
```

Plot Normal

```python
# Normal
from scipy.stats import norm

#x-axis ranges from -7 and 5 with .001 steps
x = np.arange(-7, 7, 0.001)

 
#plot normal distribution with mean   and standard deviation  
plt.plot(x, norm.pdf(x, 0, 1))
plt.plot(x, norm.pdf(x, 0, 2))
plt.plot(x, norm.pdf(x, 1, 1))

#add axes labels and a title
plt.ylabel('Density')
plt.xlabel('x')
plt.title('Normal Distribution', fontsize=14)

plt.show()
```

Binomial

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import seaborn as sns
```

```python
 # Set parameters
n = 25
p = 0.4
```

```python
# Generate random samples
x = np.random.binomial(n, p, size=1000)
print(x[:10]) 
```

```python
# Plot histogram of samples
sns.histplot(x,  kde=False, label='Samples')
```

```python
# Calculate theoretical pmf values
k = np.arange(0, n+1)
pmf1 = binom.pmf(k, 25, 0.4)
pmf2 = binom.pmf(k, 4, 0.1)
```

```python
# Normal
from scipy.stats import norm

#x-axis ranges from -7 and 5 with .001 steps
x = np.arange(0, 25, 1)
mu = 25*0.4
variance = 25 * 0.4*(1-0.4)
norm1 = norm.pdf(x, mu, np.sqrt(variance))

 
```

```python
# Plot pmf values as line graph
plt.plot(k, pmf1, 'o-', label='PMF 25-0.4')
plt.plot(k, pmf2, 'o-', label='PMF 4-0.1')
#plot normal distribution with mean   and standard deviation  
plt.plot(x, norm1, label='NORMAL 10-6')
# Add labels and legend
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Binomial distribution')
plt.legend()
```
