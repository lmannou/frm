```python
import string
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

Log normal distribution

```python
# Log normal distribution : 0, 1
from scipy.stats import lognorm
from scipy.stats import norm

mu = 0
sigma = 1

s = sigma
loc = 0
scale = math.exp(mu)

x = np.arange(0, 4, 0.001)

log_norma_mean = math.exp(mu + sigma/2)
    
print("mean of log normal is ", log_norma_mean)
 
#plot normal distribution with mean   and standard deviation  
#plt.plot(x, norm.pdf(x, 0, 1))
plt.plot(x, lognorm.pdf(x, s, loc, scale))
```

Usage of KDE (exemple)

```python
from scipy.stats import gaussian_kde
from numpy import linspace
from numpy import random

# create fake data (standard normal)
data = random.randn(1000)
# this create the kernel, given an array it will estimate the probability over that values
kde = gaussian_kde( data )
# these are the values over wich your kernel will be evaluated
dist_space = linspace( min(data) - 1, max(data) +1, 100 )
# plot the results
plt.plot( dist_space, kde(dist_space) )
```

Get N samples for the mean 
Sample date from Log normal
Calculate mean of the sample
Plot the mean distribution

```python
# get sample from lognormal
from statistics import mean

from scipy.stats import gaussian_kde
from numpy import linspace

#Get N samples for the mean

# N sample
N = 10000

for n in (10, 40, 160, 640):
    sample = []
    for x in range(N):
        rv = lognorm.rvs(s, loc, scale, n)
        #print(rv)
        m = mean(rv)
        sample.append(m)

    #Plot the PDF for the sample (N)
    kde = gaussian_kde(sample, bw_method=1.0)
    dist_space = linspace( 0, 4, 100 )
    #plt.legend()
    plt.plot( dist_space, kde.pdf(dist_space), label="sample with n="+str(n) )

plt.legend()
    
```

Poisson Histogram

```python
from scipy.stats import poisson

#lambda 
lam = 3

mu_poisson, variance_poisson = poisson.stats(lam)

print(mu_poisson, variance_poisson)

samples = poisson.rvs(lam, size=1000)

# Plot a histogram of the samples with 10 bins
plt.hist(samples, bins=10, density=True, color='green', alpha=0.5, label='Poissin sample')

# Plot the theoretical probability mass function of the Poisson distribution
x = np.arange(0, 15) # possible values of k
pmf = poisson.pmf(x, lam) # probability mass function at x
plt.plot(x, pmf, 'bo', label='Poisson pmf')

# Add labels and legend to the plot
plt.xlabel('Number of occurrences')
plt.ylabel('Probability')
plt.title('Histogram and pmf of Poisson distribution')
plt.legend()
plt.show()
```
