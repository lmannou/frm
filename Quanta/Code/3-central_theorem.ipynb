```python
import string
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

Log normal distribution

```python
# Log normal distribution : 0, 1
from scipy.stats import lognorm
from scipy.stats import norm

mu = 0
sigma = 1

s = sigma
loc = 0
scale = math.exp(mu)

x = np.arange(0, 4, 0.001)

log_normal_mean = math.exp(mu + sigma/2)
    
print("mean of log normal is ", log_normal_mean)
 
#plot normal distribution with mean   and standard deviation  
#plt.plot(x, norm.pdf(x, 0, 1))
plt.plot(x, lognorm.pdf(x, s, loc, scale))
```

Get N samples for the mean

```python
# get sample from lognormal
from statistics import mean

from scipy.stats import gaussian_kde
from numpy import linspace

#Get N samples for the mean

# N sample
N = 10000

for n in (10, 40, 160, 640):
    sample = []
    for x in range(N):
        rv = lognorm.rvs(s, loc, scale, n)
        #print(rv)
        m = mean(rv)
        # SQRT(n) * (mean estimator - mu)
        m = math.sqrt(n) * (m - log_normal_mean)
        #m =  (m - log_normal_mean)

        sample.append(m)
    #print(mean(sample))
    #Plot the PDF for the sample (N)
    kde = gaussian_kde(sample, bw_method=1.0)

    #print(kde(0))
    dist_space = linspace( -10, 10, 100 )
    
    plt.plot( dist_space, kde.pdf(dist_space), label="sample with n="+str(n) )
plt.legend(loc='best')

    
```

Poisson

```python

from scipy.stats import poisson

#lambda 
lam = 3
mu_poisson, variance_poisson  = poisson.stats(lam)

print(mu_poisson, variance_poisson)
sigma_poisson = math.sqrt(variance_poisson)
print('sigma=', sigma_poisson)
x = np.arange(0, 25, 1)
poisson_pmf1 = poisson.pmf(mu_poisson, x)
plt.plot(x, poisson_pmf1, 'o-', label='PMF POISSON '+str(mu_poisson))
plt.legend()

 
```
