```python
import string
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def portfolio_variance(sigma1, sigma2, sigma12, w):
  return np.square(w) * np.square(sigma1) + np.square(1-w)*np.square(sigma2) + 2*w*(1-w)*sigma12
```

```python
def portfolio_variance_min(sigma1, sigma2, sigma12):
  return(np.square(sigma2) - sigma12)/(np.square(sigma1) + np.square(sigma2) - 2*sigma12)
```

```python
def correlation(sigma1, sigma2, sigma12):
  return sigma12 / (sigma1, sigma2)
```

```python
def portfolio_variance_2(sigma1, sigma2, sigma12):
    w = portfolio_variance_min(sigma1, sigma2, sigma12)
    return portfolio_variance(sigma1, sigma2, sigma12, w)
```

```python
def portfolio_variance_3(sigma1, sigma2, correlation):
    sigma12 = correlation * sigma1 * sigma2
    return  portfolio_variance_2(sigma1, sigma2, sigma12)
```

```python
def portfolio_std(sigma1, sigma2, correlation):
    return np.sqrt(portfolio_variance_3(sigma1, sigma2, correlation))
```

```python
sigma1 = 0.2
sigma2 = 0.1
correlation = -0.8

```

```python
print(portfolio_std(sigma1, sigma2, correlation))
```
